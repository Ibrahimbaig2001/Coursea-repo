```python
import pandas as pd
```

```python
df = pd.read_csv("house_price.csv")
```

```python
df.head() # Printing the head of the dataset , first 5 elements
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(df.area,df.price,color = 'red',marker = '*')
plt.xlabel('Area of House')
plt.ylabel("Price of the huse in lakhs")
plt.title("Finding the relationship b/w Area and Price")
```

**Model** **Building**

```python
from sklearn import linear_model # importing linear model from skikit learn library
reg = linear_model.LinearRegression() # making reg as linearregression
```

```python
# training the data y = mx+b coeff and y-intercept, building a linear model
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3600]]) # predicting a price for the entered area
```

```python
669452-680000
# Calculating the error made by the model during prediction by differentiating the predicted by with the original price
```

```python
reg.coef_ # calculating x coefficient
```

```python
reg.intercept_ # calculating y-intercept
```

```python
135.78767123 * 2100 + 180616.43835616432 # Calculating manually by substituting the values in the y=mx+b equation
```

```python
reg.predict([[4000]])
```

```python
135.78767123 * 4000 + 180616.43835616432 # calculating manually to know the difference between prediction of the model and the manual one
```
